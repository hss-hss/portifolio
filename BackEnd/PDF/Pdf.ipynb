```python
import PyPDF2 as pyf 
from pathlib import Path

Path("Paginas").mkdir(exist_ok=True)
nome = "primeiro.pdf"

with open(nome, "rb") as arquivo:
    leitor = pyf.PdfReader(arquivo)

    for i, pagina in enumerate(leitor.pages):
        num_pagina = i + 1
        escritor = pyf.PdfWriter()
        escritor.add_page(pagina)

        with open(f"paginas/arquivo{num_pagina}.pdf", "wb") as novo_pdf:
            escritor.write(novo_pdf)
```

```python
num_paginas = [1,2,3]
novo_arquivo = pyf.PdfWriter()

for num in num_paginas:
    leitor = pyf.PdfReader(f"paginas/Arquivo{num}.pdf")
    novo_arquivo.add_page(leitor.pages[0])
    
with Path("Consolidado.pdf").open('wb') as arquivo:
    novo_arquivo.write(arquivo)
```

```python
pdf_mesclado = pyf.PdfMerger()

pdf_mesclado.append('primeiro.pdf')
pdf_mesclado.append('segundo.pdf')

with Path(f"mesclado.pdf").open(mode="wb") as arquivo:
    pdf_mesclado.write(arquivo)
```

```python
pdf_mesclado = pyf.PdfMerger()

pdf_mesclado.append('segundo.pdf')
pdf_mesclado.append('paginas/arquivo1.pdf')

with Path(f"relatorio 2 trimestes.pdf").open(mode="wb") as arquivo:
    pdf_mesclado.write(arquivo)
```

```python
arquivo_pdf_original = pyf.PdfReader('primeiro.pdf')

novo_arquivo = pyf.PdfWriter()

for pagina in arquivo_pdf_original.pages:
    pagina.rotate(90)
    novo_arquivo.add_page(pagina)
    
    with Path(f"Paginas Rotacionadas.pdf").open(mode='wb') as arquivo:
        novo_arquivo.write(arquivo)
```

```python
arquivo = pyf.PdfReader("primeiro.pdf")
qtde_paginas= len(arquivo.pages)
print(qtde_paginas)

metadados_arquivo = arquivo.metadata
print(metadados_arquivo)
```

```python
texto_referencia = "compras"

for i, pagina in enumerate(arquivo.pages):
    texto_pagina = pagina.extract_text()
    if texto_referencia in texto_pagina:
        print("Numero Pagina: ", i+1)
        texto_analisar = texto_pagina
```

```python
posicao_inicial = texto_analisar.find(texto_referencia)
posicao_final = texto_analisar.find("|", posicao_inicial + 1)

if posicao_inicial != -1 and posicao_inicial != -1:
    texto_final = texto_analisar[posicao_inicial:posicao_final]
    print(texto_final)
else: 
    print("Texto de referência não encontrado.")
```

```python
from pikepdf import Pdf, PdfImage

Path("imagens").mkdir(exist_ok=True)
img_pdf = Pdf.open('primeiro.pdf')
for pagina in img_pdf.pages:
    for nome, imagem in pagina.images.items():
        imagem_salvar = PdfImage(imagem)
        imagem_salvar.extract_to(fileprefix=f"imagens/{nome}")
```
