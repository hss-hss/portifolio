```python
import PyPDF2 as pyf 
from pathlib import Path

Path("Paginas_2").mkdir(exist_ok=True)
nome = "Senai.pdf"

with open(nome, "rb") as arquivo:
    leitor = pyf.PdfReader(arquivo)

    for i, pagina in enumerate(leitor.pages):
        num_pagina = i + 1
        escritor = pyf.PdfWriter()
        escritor.add_page(pagina)

        with open(f"paginas{num_pagina}.pdf", "wb") as novo_pdf:
            escritor.write(novo_pdf)
```
