```python
import matplotlib.pyplot as plt
meses = [ 'Jan', 'Fev', 'Mar', 'Abr', 'Mai']
vendas= [54,60,78,72,88]
plt.plot(meses, vendas)
plt.xlabel('Mês')
plt.ylabel('Vendas')
plt.title('Crescimento das Vendas ao Longo de cinco meses')
plt.grid(True)
plt.show()
```

```python
turma = ['1°EM', '2°EM', '3°EM']
qtde_alunos = [25,20,27]
barras = plt.bar(turma, qtde_alunos,color = 'blue' )
plt.bar_label(barras,labels=qtde_alunos,padding=3,fontsize=9,fontweight='bold')
plt.xlabel('Turma')
plt.ylabel('Quantidade de Alunos')
plt.title('Comparação Entre Salas do Ensino Médio')
plt.show()
```

```python
produtos = ['Arroz', 'Bolachas', 'Sabão', 'Refigerante']
quantidades = [40,52,15,45]
descontrole = [0.1,0,0,0]

cores = plt.get_cmap('primary1').colors

plt.pie(quantidades,labels=produtos, colors=cores, autopct='%.1f',)
plt.title('Produtos Mais Vendidos do Mercado')
plt.show()
```
