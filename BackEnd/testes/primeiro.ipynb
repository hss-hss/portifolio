```python
imposto = 0.3

def preco_imposto(preco):
    return preco * (1 + 0.3)
print(preco_imposto(100))
```

```python
preco_imposto2 = lambda preco: preco * (1 + imposto)
print(preco_imposto2(100))
```

```python
preco_tech = {'notebook asus': 2450, 'iphone':4500, 'samsung galaxy':3000,'tv samsung': 1000,'ps5':3000, 'tablet':1000, 'notebook dell':3000, 'ipad':3000, 'tv philco':800}

print(preco_tech)
```

```python
def ehmaior2000(item):
    return item[1] >  2000

produtos_acima2000 = dict(list(filter(ehmaior2000,preco_tech.items())))
print(produtos_acima2000)

prodto2_acia2000 = dict(list(filter(lambda item : item[1]>2000, preco_tech.items())))

print(produtos_acima2000)
print(prodto2_acia2000)
```

```python
def calcular_impostos(imposto):
    return lambda preco:preco *(1 + imposto)

calcular_preco_produto = calcular_impostos(0.1)
calcular_preco_servico = calcular_impostos(0.15)
calcular_preco_royalites = calcular_impostos(0.25)

print((calcular_preco_produto(100)))
print((calcular_preco_servico(100)))
print((calcular_preco_royalites(100)))
```
