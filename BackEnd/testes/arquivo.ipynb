```python
import PyPDF2 as pyf
from pathlib import Path

Path("paginas").mkdir(exist_ok=True)
nome = "primeiro.pdf"
with open(nome, "rb") as arquivo:
    leitor = pyf.PdfFileReader(arquivo)
    for i, pagina in enumerate(leitor.pages):
        num_pagina = i + 1

        escritor = pyf.PdfFileWriter()
        escritor.add_page(pagina)

        with open(f"paginas/Arquivo {num_pagina}.pdf", "wb") as novo_pdf:
            escritor.write(novo_pdf)
```
