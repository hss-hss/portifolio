```python
import matplotlib.pyplot as plt

meses = ['Jan','Fev','Mar','Abr','Mai','Jun']
vendas = [1500,1800,1700,2000,2200,2100]

fig, axs = plt.subplots(nrows=1,ncols=2,figsize=(8,4))

axs[0].plot(meses, vendas)
axs[1].bar(meses, vendas)
plt.show()
```

```python
alunos = ['Joao','Maria','Henrique','Eliás','Debora']
notas = [9,10,7,5,8.5]

fig, axs = plt.subplots(nrows=1,ncols=2,figsize=(12,6))

axs[0].bar(alunos,notas)
axs[0].set_xlabel("Alunos")
axs[0].set_ylabel("Notas")
axs[0].set_title("Notas Dos Alunos")

axs[1].pie(notas,labels = alunos,autopct='%1.0f')
axs[1].set_title('Comparação entre as maiores notas da sala')

plt.show()
```

```python
funcionarios = ['Joao','Maria','Henrique','Eliás']
vendas_mes = [65,79,82,91] 

fig, axs = plt.subplots(nrows=1,ncols=2,figsize=(12,6))

axs[0].barh(funcionarios,vendas_mes)
axs[0].set_ylabel("funcionarios")
axs[0].set_xlabel("Vendas no Mês")
axs[0].set_title("Grafico de Produção de Funcionários")

axs[1].bar(funcionarios,vendas_mes)
axs[1].set_ylabel("Vendas no Mês")
axs[1].set_xlabel("funcionarios")
axs[1].set_title("Grafico de Produção de Funcionários")
plt.show()
```

```python
meses = ['Jan','Fev','Mar','Abr','Mai','Jun']
gastos = [2100,3000,2000,2500,3600,4000]

fig, axs = plt.subplots(nrows=1,ncols=2,figsize=(12,6))
axs[1].pie(gastos,labels = meses,autopct='%1.0f')
axs[1].set_title('Comparação de gastos')

axs[0].bar(meses,gastos)
axs[0].set_xlabel("meses")
axs[0].set_ylabel("gastos")
axs[0].set_title("Gastos no ano")



plt.show()
```

```python
nomes = ['Cristina','Felipe','Helio','João','Naruto','Edward']
alturas = [1.60,1.75,1.82,1.70,1.90,2.0]

fig, axs = plt.subplots(nrows=1,ncols=2,figsize=(10,4))

axs[0].scatter(nomes, alturas)
axs[0].set_ylabel('Altura')
axs[0].set_xlabel('Nomes')
axs[0].set_title('Comparação de diferentes altura')
axs[0].grid(True)

axs[1].bar(nomes,alturas)
axs[1].set_ylabel('Altura')
axs[1].set_xlabel('Nomes')
axs[1].set_title('Comparação de diferentes altura')
plt.show()
```

```python
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(12, 4))

prod1 = [80,120,130,100]
prod2 = [100,120,110,130]
meses = ['Jan','Fev','Mar','Abr']

cores = plt.get_cmap('Pastel2').colors

axs[0].plot(meses,prod1, label = 'Produto 1',color=cores[3])
axs[0].plot(meses,prod2, label = 'Produto 2')
axs[1].bar(meses, prod1, label='Arroz',color=cores[5])
axs[1].bar(meses, prod2, bottom=prod1, label='Feijão',color=cores[2])
axs[1].legend()

plt.show()
```

```python
notas = (4,5,6, 7, 8, 9, 10)
frequencia =(2,9,9,1,3,1,6)
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(12, 4))

axs[0].hist(notas, bins=6, edgecolor='black')
axs[0].set_title('histograma das notas')
axs[0].set_xlabel('notas')

axs[1].bar(notas, frequencia)
axs[1].set_title('Distribuição das Notas (Barras)')
axs[1].set_xlabel('Notas')


plt.show()
```
