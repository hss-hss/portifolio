```python
import pandas as pd

vendas_df = pd.read_csv(r'Vendas.csv',sep = ';')
produtos_df = pd.read_csv(r'CadastroProdutos.csv', sep = ';')
lojas_df = pd.read_csv(r'Lojas.csv', sep = ';')
clientes_df = pd.read_csv(r'Clientes.csv', sep = ';')

display(vendas_df)
display(produtos_df)
display(lojas_df)
display(clientes_df)
```

```python
colunasParaRemover = ['Unnamed: 7', 'Unnamed: 8', 'Unnamed: 9', 'Unnamed: 10']
clientes_df = clientes_df.drop(columns=[coluna for coluna in colunasParaRemover if coluna in clientes_df.columns])
display(clientes_df)
```

```python
colunasParaRemover = ['Unnamed: 7', 'Unnamed: 8', 'Unnamed: 9', 'Unnamed: 10']
produtos_df = produtos_df.drop(columns=[coluna for coluna in colunasParaRemover if coluna in clientes_df.columns])

display(produtos_df)
```

```python
vendas_df = vendas_df.rename(columns={'E-mail':'E-mail Do Cliente'})
display(vendas_df)
```

```python
frequencia_clientes=vendas_df['E-mail Do Cliente'].value_counts()
display(frequencia_clientes)
frequencia_clientes[:5].plot()
```

```python
vendas_df = vendas_df.merge(produtos_df, on="ID Produto")
vendas_df = vendas_df.merge(lojas_df, on="ID Loja")
vendas_df = vendas_df.merge(clientes_df, on= "ID Cliente")

display(vendas_df)
```

```python
frequencia_clientes=vendas_df['E-mail Do Cliente'].value_counts()
frequencia_clientes[:5].plot(figsize=(15, 5))
```

```python
vendas_lojas = vendas_df.groupby('Nome da Loja').sum()
vendas_lojas = vendas_lojas[['Quantidade Vendida', 'Quantidade Devolvida']]
display(vendas_lojas)
```
